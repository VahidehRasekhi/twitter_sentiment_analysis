```python
import pandas as pd
import numpy as np
import time
import json
```

```python
world_cities=pd.read_csv('world-cities_csv.txt', sep=",")
```

```python
world_cities
```

```python
world_cities=world_cities.drop(columns=['geonameid'])
```

```python
world_cities['country'].unique()
```

```python
world_cities_pop=pd.read_csv('worldcities.csv')
```

```python
world_cities_pop=world_cities_pop[['city', 'country', 'population']]
```

```python
world_cities_pop=world_cities_pop.merge(world_cities, left_on= ['city', 'country'], right_on=['name', 'country'], how='left').\
                                drop(columns=['name', 'geonameid'])
```

```python
world_cities_pop[world_cities_pop['city']=='California']
```

```python
world_cities[world_cities['country']=='United States']
```

```python
world_cities_pop['location']=world_cities_pop['city']
```

```python
world_cities_pop
```

```python
world_countries_pop=world_cities_pop.groupby('country')['population'].sum().to_frame().reset_index()
```

```python
world_cities_pop=world_cities_pop[world_cities_pop['population'].notna()]
```

```python
world_cities_pop=world_cities_pop.loc[world_cities_pop.groupby('city')['population'].idxmax()]
```

```python
world_cities_pop
```

```python
world_cities_pop[world_cities_pop['city']=='London']
```

```python
world_countries_pop['location']=world_countries_pop['country']
```

```python
world_countries_pop
```

```python
US_states=world_cities_pop[world_cities_pop['country']=='United States'].groupby(['subcountry', 'country'])['population'].sum().to_frame().reset_index()
```

```python
US_states['location']=US_states['subcountry']
```

```python
US_states
```

```python
population=pd.concat([US_states, world_countries_pop, world_cities_pop], ignore_index=True)
population
```

```python
population['location']=population['subcountry']
```

```python
population=population.loc[population.groupby('location')['population'].idxmax()]
```
