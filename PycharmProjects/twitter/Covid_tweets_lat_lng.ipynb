```python
import pandas as pd
import numpy as np
import json
```

```python
with open('covid_tweets_final_df.json', 'r') as myfile:
    data=myfile.read()
    
obj= json.loads(data)
covid_tweets_analysis=pd.DataFrame(obj)
```

```python
covid_tweets_analysis
```

```python
world_cities=pd.read_csv('worldcities.csv')
world_cities
```

```python
world_cities=world_cities.drop(columns=['city_ascii','iso2', 'iso3','admin_name','capital', 'id'])
```

```python
world_cities = world_cities.rename(columns={'lat': 'Latitude', 'lng': 'Longitude'})
```

```python
world_cities
```

```python
covid_tweets=covid_tweets_analysis.merge(world_cities, left_on='location', right_on='city', how='left')
covid_tweets
```

```python
covid_tweets=covid_tweets.drop(columns=['city', 'country_y', 'population_y'])
```

```python
covid_tweets = covid_tweets.rename(columns={'country_x': 'country', 'population_x': 'population'})
```

```python
covid_tweets
```

```python
covid_tweets.to_csv('covid_tweets_lat_lng.csv')
```
