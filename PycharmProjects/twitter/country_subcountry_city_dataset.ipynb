```python
total_country=pivoted_data.groupby(['country', 'country_code']).agg({'population':'max', 'neg':'sum', 'pos':'sum'}).reset_index()
total_country
```

```python
total_country=pd.concat([total_country, other_location_sentiment])
total_country
```

```python
total_subcountry=pivoted_data.groupby(['subcountry', 'country', 'country_code']).agg({'population':'max', 'neg':'sum', 'pos':'sum'}).reset_index()
total_subcountry
```

```python
total_city=pivoted_data.dropna(subset=['city'])
total_city
```

```python
for df in (total_country, total_city, total_subcountry):
    df['total_tweets']=df['neg']+ df['pos']
    df['pos_ratio']=df['pos']/df['total_tweets']
```
